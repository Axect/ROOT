## 2. Graph

```c++
const Int_t n = 1000;
Double_t x_1[n], y_1[n], x_2[n], y_2[n];

const Double_t h = 0.01;

for (Int_t i=0; i < n; ++i) {
    x_1[i] = h * i;
    x_2[i] = h * i;
    
    y_1[i] = sin(x_1[i]);
    y_2[i] = cos(x_2[i]);
}

TGraph *tg1 = new TGraph(n, x_1, y_1);
TGraph *tg2 = new TGraph(n, x_2, y_2);

TCanvas *tC2 = new TCanvas("tC2", "graph", 1000, 600);

tC2->cd(1);
tC2->SetGrid();
tg1->Draw("AC");
tg2->Draw("C");
tC2->Draw();
```
